## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os, subprocess
import pandas as pd

REPO_URL = "https://github.com/rajnarottam38-tech/Starter_Notebooks"

if not os.path.exists("Starter_Notebooks"):
    subprocess.run(["git", "clone", REPO_URL])

os.chdir("Starter_Notebooks")

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Number 1: correlation between word count and avg_position (lower position = better rank)
print("Correlation word_count vs avg_position:", round(df["word_count"].corr(df["avg_position"]), 3))

# Number 2: average CTR for top-ranked vs lower-ranked pages
top_ranked = df[df["avg_position"] <= 10]["ctr"].mean()
lower_ranked = df[df["avg_position"] > 10]["ctr"].mean()
print("Avg CTR top10:", round(top_ranked, 3), "| Avg CTR below top10:", round(lower_ranked, 3))

# Number 3: how many pages have HIGH competition
print("HIGH competition pages:", (df["competition_level"] == "HIGH").sum(), "/", df.shape[0])
```
